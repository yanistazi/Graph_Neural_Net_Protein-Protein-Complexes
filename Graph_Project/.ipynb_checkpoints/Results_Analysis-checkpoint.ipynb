# Create a plot function customizable

```python
def plot_losses(paths=[None],nrow=1,ncol=3,xstarts=None,xends=None,ylims=None):
    fig, axs = plt.subplots(nrow, ncol,figsize=(6*ncol,6*nrow))
    for i in range(len(paths)):
        with open (paths[i]+"train_loss.txt","rb") as fp:
            list_train=[float(x.decode('utf8').strip()) for x in fp.readlines()]
        with open (paths[i]+"valid_loss.txt","rb") as fp:
            list_valid=[float(x.decode('utf8').strip()) for x in fp.readlines()]

        xstart = xstarts[i] if xstarts!=None else 10
        xend = xends[i] if xends!=None else len(list_valid)
            
        axs[i].plot(range(xstart, xend), list_valid[xstart:xend],
                 c='red', label='valid') 
        axs[i].plot(range(xstart, xend), list_train[xstart:xend],
                 c='blue', label='train')   

        axs[i].set_xlabel("Number of epoch")
        axs[i].legend()
        if ylims!=None :
            axs[i].set_ylim(0,ylims[i])    
    plt.show()
        
```

## 1. Plot test loss over different training iterations

```python
import ast
import numpy as np
indices_4 = [10,122,195,323,433,533,645,733]
pretrained_4 = ["latest_gridsearch_irmsd_5/random_split_test_3PC8_3K75/4/testing_results_"+
              str(ind)+"_" for ind in indices_4]
pretrained_gridsearchs = [pretrained_4]
mean_abs_err = []

max_normalized = 20
min_normalized = 0
cap = 10

for pretrained_gridsearch in pretrained_gridsearchs:
    print(pretrained_gridsearch[1].split("/")[-2])
    for pretrain in pretrained_gridsearch :
        all_pred = []
        all_y = []
        all_abs_list = []
        for complex in ["3K75","3PC8","3hmx"]:#,"3hmx"]:
            test_data = h5py.File(pretrain+complex+"/test_data.hdf5")
            list_mols = list(test_data["epoch_0000"]["test"]["mol"])
            list_targets = list(test_data["epoch_0000"]["test"]["targets"])
            list_outputs = list(test_data["epoch_0000"]["test"]["outputs"])
            test_data.close()
            
            list_y_raw = [l*max_normalized for l in list_targets]
            list_pred_raw = [l*max_normalized for l in list_outputs]
            list_pred = [ min_normalized if l<min_normalized else cap if l>cap else l for l in list_pred_raw ]
            list_y = [ min_normalized if l<min_normalized else cap if l>cap else l for l in list_y_raw ]
            all_pred.append(list_pred)
            all_y.append(list_y)
            abs_list = [abs(l1-l2) for l1,l2 in zip(list_y,list_pred)]
            all_abs_list.append(abs_list)
        mean_abs_err.append(np.mean(list(itertools.chain(*all_abs_list))))
        
        
plt.figure(figsize=(10,4))
sns.lineplot(np.array([10,122,195,323,433,533,645,733]),
         np.array(mean_abs_err),marker="o")
plt.show()        



```

# Find good and bad examples from selected model

```python
import h5py
test_data = h5py.File("/bgfs01/insite/yanis.tazi/Scoring_Project/Graph_Project/\
latest_gridsearch_irmsd_5/random_split_test_3PC8_3K75/4/testing_results_733_3hmx/test_data.hdf5")
list_mols = list(test_data["epoch_0000"]["test"]["mol"])
list_targets = list(test_data["epoch_0000"]["test"]["targets"])
list_outputs = list(test_data["epoch_0000"]["test"]["outputs"])
test_data.close()
```

```python
list_targets_raw = [l*20 for l in list_targets]
list_outputs_raw = [l*20 for l in list_outputs]
abs_list = np.array([abs(l1 - l2)*20 for l1,l2 in zip(list_targets_raw,list_outputs_raw)])

list_targets_capped = [ 0 if l<0 else 10 if l>10 else l for l in list_targets_raw ]
list_outputs_capped = [ 0 if l<0 else 10 if l>10 else l for l in list_outputs_raw ]

abs_list_capped = np.array([abs(l1 - l2) for l1,l2 in zip(list_targets_capped,list_outputs_capped)])

bad_ind = np.where((abs_list_capped>8) & (abs_list_capped<10))[0]
print("Bad",bad_ind)
# 1260
print([(list_mols[i],round(list_targets_raw[i],3),round(list_outputs_raw[i],3)) for i in bad_ind])

good_ind = np.where((abs_list_capped>0) & (abs_list_capped<0.01))[0]
print("Good",good_ind)
# 1260
print([(list_mols[i],round(list_targets_raw[i],3),round(list_outputs_raw[i],3)) for i in good_ind])
```
