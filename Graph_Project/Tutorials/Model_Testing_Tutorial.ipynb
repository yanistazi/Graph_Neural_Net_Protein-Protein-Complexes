# 1. Model Testing results

![alt text](../plots_folder/help_testing.png "Title")

We can now test the model using .pth.tar saved model weights on any complex/list of complex and get a dictionary with metrics of interest.

Test model using saved .pth.tar in pretrained_model. Needs to map pretrained model with the correct neuralnet
architecture and optimizer. It takes as input a path to .hdf5 graph files for the complexes and we can 
specify a subset of complexes to be tested from that path or all the .hdf5 files if list is empty.

It returns :

    1. test.hdf5 with molecule names , outputs and predictions
    
    If get_metrics is False
    
    2. .txt file with the predictions for all complexes + target values (if get_metrics is True)
       list with the predictions for all the pdb files.
       
    3. Returns the normalized data metrics if get_metrics is True
    
    4. If compute_raw is True return .txt file with the predictions and targets
    
    5. If compute_raw is True return .txt file with the metrics

```python
dir_path="/bgfs01/insite02/yanis.tazi/data/graph_project_data/hdf5_target_5_replicates/hdf5_pdb_graphs_copy_12/"
outdir="/bgfs01/insite/yanis.tazi/Scoring_Project/Graph_Project/output_directory_tutorial/testing_tutorial/"
list_complexes = None # Specify it later
neuralnet_arch = EGAT_Net
pretrained_model = "/bgfs01/insite/yanis.tazi/Scoring_Project/Graph_Project/latest_gridsearch_irmsd_5/\
random_split_test_3PC8_3K75/4/treg_yscoring_normalized_b1024_e1000_lr0.001_733.pth.tar"
optimizer_name = "Adam"
threshold = 5
get_metrics = True
compute_raw = True
min_val = 0
max_val = 20
complexes_decoys_to_test = ["3PC8","3K75"]

# Test for multiple decoys of 3K75
for l in complexes_decoys_to_test:
    model_testing(dir_path = dir_path,
                  outdir = outdir+l+"/",
                  list_complexes = [l],
                  neuralnet_arch = neuralnet_arch,
                  pretrained_model = pretrained_model,
                  optimizer_name = optimizer_name,
                  threshold = threshold,
                  get_metrics = get_metrics,
                  compute_raw = compute_raw,
                  min_val = min_val,
                  max_val = max_val)
```
