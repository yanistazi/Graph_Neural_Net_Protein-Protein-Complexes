# PART 1 : Data Preparation for protein scoring prediction project

# STEP 1. 
## Get a dictionary with labeled data for the threshold set up:

### (one dictionary per decoy stage for each complex)


Threshold

- For each complex: 
> - For water, it0 and it1 :
>> - Compute i-RMSD for all docking models 
>> - Count number_pos of all docking models with i-RMSD < Threshold ; number_neg with i-RMSD >= Threshold
>> - If number_pos > number_neg > 0
>>> - Select n_sample = max(number_neg,min_number_pdb_per_stage) and randomly (without replacement) sample n_sample positive docking and number_neg negative docking
>> - If number_neg > number_pos > 0
>>> - Select n_sample = max(number_pos,min_number_pdb_per_stage) and randomly (without replacement) sample n_sample negative docking and number_pos positive docking
>> - If number_neg = 0 and number_pos > 0
>>> - Select randomly min_number_pdb_per_stage positive docking
>> - If number_neg > 0 and number_pos = 0
>>> - Select randomly 10negative docking
>> - Create a dictionary with docking model names and values 




#### In summary , we have now 3 dictionaries for each complex (for water, it0 and it1) with the selected pdb files based on the threshold+sampling explained above .
#### An example with complex 1JTD it1 :
- ({'1JTD_it1_complex_329.pdb': 14.98,
        '1JTD_it1_complex_359.pdb': 11.385})
        
Select pdbs based on those criterias and create a target.txt files with the name of the selected pdbs and their corresponding i-RMSD for each complex and for each stage.        

```python
dir_path = "/bgfs01/insite02/yanis.tazi/data/graph_project_data/131/"
seed = 912
threshold_irmsd = 5
min_number_pdb_per_stage = 1
list_complexes = ""
txt_file_name="target_5.txt"
max_val = 20
save = False  # set to True to save txt_file_name

create_dict_selected_pdbs(dir_path = dir_path,
                              seed = seed,
                              threshold_irmsd = threshold_irmsd,
                              min_number_pdb_per_stage = min_number_pdb_per_stage,
                              list_complexes = list_complexes,
                              txt_file_name = txt_file_name,
                              save = save,
                              max_val = max_val)
```

# Distribution of i-RMSD

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
from itertools import chain
import warnings
warnings.filterwarnings("ignore")
all_selected = []
path = "/bgfs01/insite02/yanis.tazi/data/graph_project_data/hdf5_target_5_replicates/hdf5_pdb_graphs_copy_1/"
all_selected = []
for l in os.listdir(path):
     # new to plot each
    try:
        values = np.loadtxt(path+l+"/target_list.txt", delimiter=" ", usecols=[1])
        all_selected.append(values)
    except:
        pass
all_selected = list(chain(*all_selected))
plt.figure(figsize=(6,6))
sns.distplot(all_selected,hist=False,bins=100,label="all stages selected PDB",color="gray")
plt.legend()
plt.xlim(0,20)
# plt.ylim(0,0.5)
plt.text(12,0.15,"Total Count="+str(len(all_selected)), fontsize=12)
plt.text(10,0.12,"Count (i-RMSD <5)="+str(len([l for l in all_selected if l < 5])), fontsize=12)
plt.text(15,0.09,"min="+str(min(all_selected)), fontsize=12)
plt.text(15,0.06,"mean="+str(np.round(np.mean(all_selected),3)), fontsize=12)
plt.text(15,0.03,"max="+str(max(all_selected)), fontsize=12)
plt.legend(fontsize=12)
plt.xlabel("i-RMSD", fontsize=18)
plt.ylabel("Density", fontsize=18)

plt.savefig("../plots_folder/Density_plots_selected.png")
plt.show()



for l in os.listdir(path):
    all_selected = []
     # new to plot each
    try:
        values = np.loadtxt(path+l+"/target_list.txt", delimiter=" ", usecols=[1])
        all_selected.append(values)
    except:
        pass
    all_selected = list(chain(*all_selected))
    sns.distplot(all_selected,hist=True,bins=100,label="all stages selected PDB " + l,color="lightblue")
    plt.legend()
    plt.xlim(0,20)
    # plt.ylim(0,0.5)
    plt.text(12,0.23,"Total Count="+str(len(all_selected)), fontsize=12)
    plt.text(15,0.18,"min="+str(min(all_selected)), fontsize=12)
    plt.text(15,0.14,"mean="+str(np.round(np.mean(all_selected),3)), fontsize=12)
    plt.text(15,0.09,"max="+str(max(all_selected)), fontsize=12)
    plt.legend(fontsize=12)
    plt.xlabel("i-RMSD", fontsize=18)
    plt.ylabel("Density", fontsize=18)

#     plt.savefig("../plots_folder/Density_plots_selected.png")
    plt.show()
```
